```python
def name_cleaner(series):
    final_series = []
    for name in series:
        if "*" in name:
            name_list = name.split('*')
            final_series.append(name_list[0])
        elif "\\" in name:
            name_list = name.split("\\")
            final_series.append(name_list[0])
    return np.array(final_series)
```

```python
def normalize_features(df):
    mu = df.mean()
    sigma = df.std()
    if (sigma == 0).any():
        raise Exception("One or more features had the same value for all samples, and thus could " + \
                         "not be normalized. Please do not include features with only a single value " + \
                         "in your model.")
    df_normalized = (df - df.mean())/df.std()
    return df_normalized, mu, sigma
```

```python
def compute_cost(features, values, theta):
    m = len(values)
    sum_squared_errors = ((np.dot(features, theta) - values)**2).sum()
    cost = sum_squared_errors/(2*m)
    return cost
```

```python
def gradient_descent(features, values, theta, alpha, num_iterations):
    m = len(values)
    cost_history = []
    for i in range(num_iterations):
        predicted_values = np.dot(features,theta)
        theta = theta + alpha/m * (np.dot(values - predicted_values, features))
        cost = compute_cost(features,values,theta)
        cost_history.append(cost)
    return theta, pd.Series(cost_history)
```

```python
def predictions(dataframe):
    features = dataframe[dataframe["GS_last"] > 3][["Tgt_last","Rec_last","cYds_last"]]
    values = dataframe[dataframe["GS_last"] > 3]["cYds"]
    features, mu, sigma = normalize_features(features)
    m = len(values)
    features['ones'] = np.ones(m)
    alpha = .1
    num_iterations = 200
    theta_gradient_descent = np.zeros(len(features.columns))
    theta_gradient_descent, cost_history = gradient_descent(features, 
                                                            values, 
                                                            theta_gradient_descent, 
                                                            alpha, 
                                                            num_iterations)
    model = sm.OLS(values, features)
    results = model.fit()
    predictions = results.predict(features)
    return predictions
```

```python
def compute_r_squared(data, predictions):
    avg = np.mean(data)
    sum_squared_errors = ((data-predictions)**2).sum()
    denom = ((data - avg)**2).sum()
    return (1-sum_squared_errors/denom)
```
